## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

* если источник traffic_source равен yandex или google, то в source_type ставится organic
* для источников paid и email из России - ставим ad
* для источников paid и email не из России - ставим other
* все остальные варианты берем из traffic_source без изменений


```python
import pandas as pd
```

```python
df = pd.read_csv('visit_log.csv', sep = ';')
```

```python
def get_source_type(row):
    traffic_source = row['traffic_source']
    region = row['region']
    result = ''
    if (traffic_source == 'yandex' ) | (traffic_source == 'google' ):
        result = 'organic'
    elif (traffic_source == 'paid' ) | (traffic_source == 'email' ):
        if region == 'Russia':
            result = 'ad'
        else:
            result = 'other'
    else:
        result = traffic_source
    return result
```

```python
df['source_type'] = df.apply(get_source_type, axis = 1)
df.head(50)
```

## Задание 3

В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
data.head()
```

```python
pip install pymystem3
```

```python
from pymystem3 import Mystem
```

```python
m = Mystem()
lemmas = m.lemmatize(data.keyword[0])
lemmas
```

```python
def get_lemmas(row):
    m = Mystem()
    lemmas = m.lemmatize(row)
    for word in lemmas:
        if (word == ' ') | (word == '\n'):
            lemmas.remove(word)
    return ' '.join(lemmas)
```

```python
get_lemmas(data.keyword[0])
```

```python
data['lemmas'] = data.keyword.apply(get_lemmas)
```
