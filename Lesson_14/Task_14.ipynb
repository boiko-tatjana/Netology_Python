## Задача 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
df = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
df['date'] = pd.to_datetime(df['timestamp'], unit='s')
```

```python
df2 = df.groupby(df.userId).date.agg(ratings_count ='count', min_date = 'min', max_date = 'max').reset_index()
```

```python
df2[df2.ratings_count >= 100]
```

```python
df2['diff'] = df2.max_date - df2.min_date
```

```python
df2.sort_values('diff')
```

```python
df2.mean()[['diff']]
```

## Задание 3

В задаче сквозной аналитики вам предоставили данные по местоположению пользователей.  
Т. е. для каждого user_id известна последовательность координат (широта/долгота), когда они требовались приложению для полноценной работы.  
Как бы вы добавили эти сведения в таблицу визитов и покупок?  
Для составления ответа можно использовать вопросы:
* У каждого пользователя известен набор координат. А для связывания с визитом или фактом покупки скорее всего потребуется одно-два числа. Как их получить?
* Наборы координат одного и того же пользователя могут быть значительно удалены друг от друга. Как это отразится на вопросе расчетах пункта 1?
* Какие дополнительные признаки можно получить из координат? Ведь это просто числа, которые сами по себе мало что дают.

В обе таблицы (с координатами и покупками) целесообразно добавить временную метку, соответвующую измерению координат или совершению покупки. 
Скорее всего время, в которое измерялась координата и совершалась покупка, могут не совпасть. Поэтому целесообразно добавить еще один столбец, который можно было бы назвать "уникальный идентификатор локации".
В таблице с координатами это может быть обычное целочисленное значение, увеличивающееся для каждого пользователя. А в таблице с покупками в качестве такого значения будет значение из таблицы с координатами при условии, что временной интервал между метками времени в таблицах будет минимальным.

```python
coords = pd.DataFrame(
    {
        'coords_user_id' : [1, 1],
        'x': [0, 5],
        'y': [0, 5],
        'coords_date' : ['11-06-2020 12:00:00', '11-06-2020 12:05:00'],
        'location_id' : [1, 2]
    }
)
coords
```

```python
sales =  pd.DataFrame(
    {
        'sales_user_id' : [1, 1],
        'goods': ['Эскимо', 'Газировка'],
        'sales_date' : ['11-06-2020 12:02:00', '11-06-2020 12:04:00'],
        'location_id' : [1, 2]
    }
)
sales
```

По столбцу 'location_id' возможно объединить таблицы.

```python
df_merge = coords.merge(sales, how = 'left', on = 'location_id')
```

В таблицу coords кроме того можно добавить два вычисляемых поля: 
* dt (разность между текущей и предыдущей временными метками);
* dr (расстояние между текущим и предыдущим местоположениями, заданными с помощью радиус-вектора).

При условии уникальности временных меток dt будет всегда отлично от 0.
Тогда для всех записей, начиная со второй, можно посчитать скорость перемещения данного человека.

По полю dr можно определить, какие покупки человек совершал в близких и далеких друг от друга локациях.
Если разделить dr на dt, то можно определить скорость с которой человек перемещался и приблизительно установить способ перемещения:
* не перемещался вообще;
* перемещался пешком;
* бежал;
* ехал на велосипеде или электросамокате;
* ехал на общественном транспорте или автомобиле.
