### Задание 2: 
Посчитайте частные производные функции в точкe
(на бумажке или в LaTeX и на numpy):

$f(x,y)=2x^2y^3 + 1/x + y^2x + 7$ в точке $(1,2)$

Решение:
1. Найдем частные производные:  

$\frac{\partial f(x,y)}{\partial x} = 2\cdot 2 \cdot x \cdot y^3 - \frac {1}{x^2} + y^2 $  

$\frac{\partial f(x,y)}{\partial y} = 2 \cdot 3 \cdot x^2 \cdot y^2 + 2\cdot y \cdot x $
2. Найдем значение частных производных в точке (1,2): 

$\frac{\partial f(1,2)}{\partial x} = 32 - 1 + 4 = 35 $  

$\frac{\partial f(1,2)}{\partial y} = 24 + 4 = 28 $

```python
vector = (1 , 2)
```

```python
def function(vector):
    x = vector[0]
    y = vector[1]
    f = 2 * x**2 * y**3 + 1/x + y**2 * x + 7
    return f
```

```python
eps = np.sqrt(np.finfo(float).eps)
approx_fprime(vector, function, [eps, eps])
```

### Задание 3

Найти точки минимума в следующих данных (колонка height):

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
data = pd.read_csv('video.csv')
data.head()
```

```python
plt.figure(figsize = (20, 5))
plt.plot(data.height)
plt.xlabel('время')
plt.ylabel('изменение высоты по времени')
```

```python
data['inv_height'] = data.height * (-1)
```

```python
peaks, _ = find_peaks(data.inv_height, prominence = 0.5, width = 80)
plt.figure(figsize = (20, 5))
plt.plot(data.inv_height)
plt.plot(peaks, data.inv_height[peaks], "x")
plt.show()
```
