Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- **p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит**;  
- **s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится**  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- **l – list – команда, которая выведет список всех документов в формате**  
**passport "2207 876234" "Василий Гупкин"**  
- **as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень.**        
*Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- **d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок.**  
*Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- **m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую.**  
*Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
# Главная пользовательская программа
def main() :
    to_continue = True
    while to_continue :
        command = input_command()
        if command == 'p' :
            print(get_person())
        elif command == 's' :
            print(get_shelf())
        elif command == 'l' :
            get_all_documents()
        elif command == 'as' :
            add_shelf()
        elif command == 'd' :
            delete_document()
        elif command == 'm' :
            move_document()
        elif command == 'a' :
            add_document() 
        elif command == 'q' :
            to_continue = False
            print('До новых встреч!')
            
        
# Вспомогательная подпрограмма для корректного ввода типа команды
def input_command() :
    correct_input = False
    while not correct_input :
        print('''Введите тип команды:
        p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;
        s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится;
        l – list – команда, которая выведет список всех документов;
        as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень;
        d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок;
        m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую;
        a – add – команда, которая добавит новый документ в каталог и в перечень полок;
        q – quit – команда, для выхода из программы.
        ''')        
        valid_commands = ['p', 's', 'l', 'as', 'd', 'm', 'a', 'q']       
        command = input()
        if len(command) > 0 :
            if command in valid_commands :
                correct_input = True
            else :
                print('Некорректная команда')
        else :
            print('Вы не ввели команду')
    return command


# Вспомогательная подпрограмма для корректного ввода номера документа
def input_document_number() :
    correct_input = False
    while not correct_input :
        document_number = input('Введите номер документа: ')
        if len(document_number) > 0 :
            correct_input = True
        else :
            print('Номер документа не может быть пустым')
    print('Вы ввели номер документа:', document_number)
    return document_number


# Вспомогательная подпрограмма для корректного ввода типа документа
def input_document_type() :
    correct_input = False
    while not correct_input :
        document_type = input('Введите тип документа: ')
        if len(document_type) > 0 :
            correct_input = True
        else :
            print('Тип документа не может быть пустым')
    print('Вы ввели тип документа:', document_type)
    return document_type


# Вспомогательная подпрограмма, предназначенная для получения
# только существующего номера документа
def input_present_document_number() :
    document_is_present = False
    while not document_is_present : 
        document_number = input_document_number()
        for number in directories.values() :
            if document_number in number :
                document_is_present = True
        if not document_is_present :
            print('Документ с указанным номером отсутствует') 
    return document_number
    

# Вспомогательная подпрограмма для корректного ввода номера полки
def input_shelf_number() :
    correct_input = False
    while not correct_input :
        shelf_number = input('Введите номер полки: ')
        if len(shelf_number) > 0 :
            if shelf_number.isnumeric() :
                shelf_number = int(shelf_number)
                correct_input = True
            else :
                print('Некорректный номер полки')
        else :
            print('Номер полки не может быть пустым')
    print('Вы ввели номер полки:', shelf_number)
    return shelf_number


# Вспомогательная подпрограмма для корректного ввода ФИ владельца
def input_person() :
    correct_input = False
    while not correct_input :
        person = input('Введите Фамилию и Имя владельца: ')
        if len(person) > 0 :
            correct_input = True
        else :
            print('Фамилия и Имя владельца не может быть пустым')
    print('Вы ввели Фамилию и Имя владельца:', person)
    return person


# Вспомогательная подпрограмма, предназначенная для получения
# только существующих номеров полок
def input_present_shelf_number() :
    shelf_is_present = False
    while not shelf_is_present :
        goal_shelf_number = input_shelf_number() 
        for  shelf_number in directories.keys() : 
            if int(shelf_number) == goal_shelf_number :
                shelf_is_present = True
                break
        if not shelf_is_present:
            print('Целевая полка с указанным номером отсутствует')  
    return goal_shelf_number


# Вспомогательная подпрограмма, которая спросит номер документа 
# и выведет имя человека, которому он принадлежит
def get_person() :
    document_number = input_document_number()
    result = 'Документ не найден'
    for docs in documents :
        if list(docs.values())[1] == document_number :
            result = list(docs.values())[2]
    return result


# Вспомогательная подпрограмма, которая спросит номер документа  
# и выведет номер полки, на которой он находится 
def get_shelf() :
    document_number = input_document_number()
    result = 'Документ не найден'
    for number in directories.items() :
        if document_number in number[1]:
            result = number[0]
    return result


# Вспомогательная подпрограмма, которая выведет список всех документов
# в заданном формате 
def get_all_documents() :
    for docs in documents :
        print(list(docs.values())[0] + ' "'
              + list(docs.values())[1] + '" ' + '"' + 
              list(docs.values())[2] + '"' )

        
# Вспомогательная подпрограмма, которая спросит номер новой полки 
# и добавит ее в перечень        
def add_shelf() : 
    shelf_number = input_shelf_number()
    
    exists = False
    
    for number in directories.keys() :
        if shelf_number == int(number) :
            print('Полка с указанным номером уже существует')
            exists = True

    if not exists :
        directories[str(shelf_number)] = []
        print('Добавление полки выполнено успешно')        
    print(directories)    


# Вспомогательная подпрограмма, которая спросит номер документа 
# и удалит его из каталога и из перечня полок     
def delete_document() :
    document_number = input_present_document_number() 
    
    for docs in documents :
        if list(docs.values())[1] == document_number :
            del documents[documents.index(docs)]
            document_is_present = True
    
    for number in directories.values() :
        if document_number in number :
            del number[number.index(document_number)]
    print('Удаление документа выполнено успешно') 
    print(documents)
    print(directories)


# Вспомогательная подпрограмма, которая спросит номер документа 
# и целевую полку и переместит его с текущей полки на целевую 
def move_document() : 
    document_number = input_present_document_number()
    goal_shelf_number = input_present_shelf_number()

    for number in directories.values() :
        if document_number in number :
            del number[number.index(document_number)]
        
    for shelf_number in directories.items() :
        if goal_shelf_number == int(shelf_number[0]) :
            shelf_number[1].append(document_number)
    print('Перемещение документа выполнено успешно')  
    print(directories)

    
# Вспомогательная подпрограмма, которая добавит новый документ в каталог  
# и в перечень полок, спросив его номер, тип, имя владельца и номер полки, 
# на котором он будет храниться 
def add_document() :
    document_type = input_document_type() 
    document_number = input_document_number()
    person = input_person()
    shelf_number = input_present_shelf_number()
    
    document = dict(type  = document_type, number = document_number,
                    name = person)
    documents.append(document)
    
    for number in directories.items() :
        if shelf_number  == int(number[0]) :
            number[1].append(document_number)
    print('Добавление документа выполнено успешно') 
    print(documents)
    print(directories)
    

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
    
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
    
main()
```
